```python
# 1. Python function example
def square_even_numbers(nums):
    result = []
    for n in nums:
        if n % 2 == 0:
            result.append(n * n)
    return result

sample_list = [1, 2, 3, 4, 5, 6]
print("Squared evens:", square_even_numbers(sample_list))

# 2. NumPy vector operation
import numpy as np
arr = np.array([10, 20, 30, 40])
print("Original:", arr)
print("Minus mean:", arr - arr.mean())

# 3. Pandas filter example
import pandas as pd
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
df = pd.read_csv(url)
filtered = df[df['age'] > 30].head()
filtered
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
df = pd.read_csv(url)

# 2. Show first 5 rows
print("Head:")
print(df.head())

# 3. Print mean age
print("\nMean age:", df['age'].mean())

# 4. Count males vs females
print("\nMale/Female counts:")
print(df['sex'].value_counts())

# 5. Plot histogram
plt.hist(df['age'], bins=20, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

### Top 10 Most Important Q&A (Week 1)

---

#### 1️⃣ What is the difference between a Python list and a NumPy array?
**Answer:**  
A Python list stores flexible data types but is slow for math.  
A NumPy array stores data in continuous memory and supports fast vectorized operations used in ML.

---

#### 2️⃣ Why do we need NumPy for machine learning?
**Answer:**  
ML relies on heavy numerical computation.  
NumPy performs fast matrix operations, supports broadcasting, and removes the need for loops.

---

#### 3️⃣ What is broadcasting in NumPy?
**Answer:**  
Broadcasting lets NumPy automatically expand smaller arrays so arithmetic can happen without loops.  
Example: `[1, 2, 3] + 10 → [11, 12, 13]`.

---

#### 4️⃣ What does axis=0 and axis=1 mean?
**Answer:**  
`axis=0` → operate down columns (vertical)  
`axis=1` → operate across rows (horizontal)

---

#### 5️⃣ What is a Pandas DataFrame?
**Answer:**  
A DataFrame is a 2D labeled table (rows + columns) used to load, clean, filter, and analyze structured data.

---

#### 6️⃣ What do head() and info() tell us?
**Answer:**  
`head()` shows the first few rows for a quick preview.  
`info()` shows column types, missing values, and structure.

---

#### 7️⃣ Difference between loc and iloc?
**Answer:**  
`loc` selects by labels (row/column names).  
`iloc` selects by integer positions (row/column index numbers).

---

#### 8️⃣ How do we handle missing values?
**Answer:**  
Option 1: Remove missing rows (`dropna`)  
Option 2: Replace values (`fillna`) — e.g., mean age  
This prevents data loss or model errors.

---

#### 9️⃣ Why is data visualization important?
**Answer:**  
Graphs help us see patterns, trends, outliers, and errors that raw tables hide — essential before modeling.

---

#### 🔟 Difference between describe() and value_counts()?
**Answer:**  
`describe()` → summary stats for numeric columns (mean, std, quartiles)  
`value_counts()` → counts of categories (e.g., male vs female)
