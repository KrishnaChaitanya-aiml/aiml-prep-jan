### Day 6 – Data Visualization for AIML

#### What I learned
- How to plot basic charts using Matplotlib
- How to visualize distributions and relationships in data
- Using plots to ask questions and spot patterns

#### What I implemented
- Histogram of passenger ages
- Bar charts of survival counts and survival by gender
- Scatter plot to compare fare against age
- Visualized missing values from the dataset

#### Observations
- A single chart reveals patterns much faster than raw numbers
- Female survival rate is significantly higher
- Fare vs age has no strong pattern but shows outliers


```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
df = pd.read_csv(url)
df
```

```python
plt.hist(df['age'], bins=[5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80, 85], edgecolor='black')
plt.title("Age Distribution on Titanic")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
from collections import Counter
status = []
people = []
survival_counts = Counter(df['survived'])
for k, v in survival_counts.items():
  status.append(k)
  people.append(v)

plt.bar(status, people, edgecolor='black')
plt.title('Titanic Survival Counts')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.tight_layout()

plt.xticks(status, ['Dead', 'Survived'])
plt.show()
```

```python
df.groupby('sex')['survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
plt.scatter(df['age'], df['fare'])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
df.isnull().sum().plot(kind="bar")
plt.title("Missing Values per Column")
plt.show()
```
