```python
import numpy as np

a = np.array([1,2,3,4,5], dtype='float32')
b = np.array([[12,14,16,18], [11,13,15,17]])
print(a)
print(b)
print(a.shape)
print(b.shape)
print(a.dtype)
print(b.dtype)
```

```python
import numpy as np

a = np.array([8,16,28,31,70,82,94,13])
print(a[0])
print(a[-1])
print(a[1:5])
b = np.array([[1,3,5,7,9],[2,4,6,8,0]])
print(b[0, :])
print(b[:, 2])
```

```python
import numpy as np

a = np.array([1,2,3,4])
b = np.array([1,0,1,0])
print(a + b)
print(a * 3)
print(a ** 2)
```

```python
import numpy as np

#Squaring digits using python loop
num = [2,4,6,8,10]
squares = []
for i in num:
    squares.append(i ** 2)
print(squares)

#Squaring digits using numpy
num = np.array([2,4,6,8,10])
print(num ** 2)
```

```python
import numpy as np

a = np.array([[17,19,16,5],[25,50,20,1]])
print(a.min())
print(a.max())
print(a.sum())
print(a.mean())
```
