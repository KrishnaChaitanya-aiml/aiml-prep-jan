```python
num = [1, 2, 3, 4, 5]
total_sum = 0
for i in num:
  total_sum = total_sum + i
print("Sum of the list is:", total_sum)
max = num[0]
for i in num:
  if i > max:
    max = i
print("Maximum of the list is:", max)
```

```python
string = "count"
vowels = "aeiou"
c = 0
for i in string:
  if i in vowels:
    c += 1
print("Number of vowels in the string is:", c)
```

```python
Dict = {'krish':100, 'Anuneeth':91, 'Vignesh':82, 'Lakshith':85}
topmarks = 0
for k, v in Dict.items():
  if v > topmarks:
    topstd = k
    topmarks = v
print(topstd,"got the highest marks:",topmarks)
```
